```python
import requests
import json
import pandas as pd
```

```python
#create a text file that contains your API key
#you DO NOT want others to see your API key
#this method is not the most secure but will work for this exercise

#this will read the file and store your API key as a variable
with open('brewDB_key.txt', 'r') as file:
    for line in file:
        api_key = line
```

```python
url = r"https://sandbox-api.brewerydb.com/v2/breweries/?key="
```

```python
#in the request to the website, give it the url and add the API key at the end
response = requests.get(url+api_key)
```

```python
response
```

```python
beerdata = response.json()
```

```python
beers = beerdata['data']
```

```python
beers[0]
```

```python
ID = []
Name = []
Est = []
Stat = []
Inbiz = []
MasOwn = []
Bio = []
Web = []
Desc = []
```

```python
btext = "NO_DATA"
for beer in beers:
    try:
        ID.append(beer['id'])
    except:
        ID.append(btext)
        
    try:
        Name.append(beer['name'])
    except:
        Name.append(btext)
        
    try:
        Est.append(beer['established'])
    except:
        Est.append(btext)
        
    try:
        Stat.append(beer['statusDisplay'])
    except:
        Stat.append(btext)
        
    try:
        Inbiz.append(beer['isInBusiness'])
    except:
        Inbiz.append(btext)
        
    try:
        MasOwn.append(beer['isMassOwned'])
    except:
        MasOwn.append(btext)
        
    try:
        Bio.append(beer['isOrganic'])
    except:
        Bio.append(btext)
        
    try:
        Web.append(beer['website'])
    except:
        Web.append(btext)
        
    try:
        Desc.append(beer['description'])
    except:
        Desc.append(btext)
```

```python
print(len(Bio), len(Desc), len(MasOwn), len(Stat), len(ID), len(Est))
```

```python
corplist = list(zip(ID, Name, Est, Stat, Inbiz, MasOwn, Bio, Web, Desc))

colnamz = ['id','name','established','StatusDisplay','isInBusiness','isMassOwned','isOrganic','website','description']
```

```python
brewer = pd.DataFrame(corplist, columns=colnamz)
brewer
```
