```python
# importing our text file, make sure it is 'read mode'
ham = ""
with open('hamlet-act1.txt', 'r') as f:
        ham = f.read()
```

```python
#looking at the text data
ham[:120]

# the output is messy and full of \n and other intrusive marks (for our purposes)
```

```python
# pretty look
print(ham[:120])
```

```python
# carving out individual words from Hamlet
list_1 = ham.split() 
print(list_1[:120]) # inspect the first 120 words or items from this list
```

```python
# let's get rid of punctuation, like that messy single quote (') just there

ham_2 = ""
punct = [",","'",".","?","!","[","]",";",'"',"-","(",")",""] # a god awful amount of punctuation--without the olde englishe
for letter in ham:
    if letter not in punct:
        ham_2 += letter
        
print(ham_2[:500]) # inspect ham_2 to the 500th current character
```

```python
# making a third data set (ham_3) that is all lowercase
ham_3 = str.lower(ham_2)
print(ham_3[:140])
```

```python
# Making a dictionary of unique words etc

list_2 = ham_3.split()
set_1 = set(list_2)
# print(set_1) # works, but is HUGE
```

```python
# making a dictionary where values are 0
dict_1 = dict.fromkeys(set_1,0)
# print(dict_1) # works, but is HUGE
```

```python
# now, updating count value

for word in list_2:
    if word in dict_1:
        dict_1[word] += 1
# print(dict_1)   # works, but is HUGE
```

```python
# looking at the dictionary list, there are 4,817 unique words in Act I! That's a large vocabulary!
len(dict_1)
```

```python
# now we are sorting by VALUE from the largest to the least
# the lowest value is 3, and there are many words there
# interesting, to be expected (the, and, of) are in the top few

# code taken from https://www.saltycrane.com/blog/2007/09/how-to-sort-python-dictionary-by-keys/

for key, value in sorted(dict_1.items(), key=lambda item: item[1], reverse=True): # reverse to show keys large to small
    print("%s: %s" % (key, value)) # PITA to get the %s working here
```
