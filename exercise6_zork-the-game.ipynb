>Zork is a famous text-only adventure game for the computer. Users would type commands and navigate the rooms of a castle.

>Write an application that asks the user what direction they wish to travel in. Once they tell you the direction, move them to the next room and tell them what is in it and what direction the other exits are.

>When you develop your program it is helpful to include each room in its own function. The role of the functions are to get input from the user and direct the user to the appropriate room.

>Your program shall allow the user to move from one room back and forth to any other connected room.

>You program should allow the user to find the secret room only 25% of the time. However, once they find the secret room they can always find it.

>When the user exits the house or quits there is a 25% chance they will be followed by a ghost. Let them know when they are being followed.

>Also let the user know how many rooms they visited after they exit or quit.

|No. | Room | Contains | Doors |
|---|---|---|---|
|#1|foyer| dead scorpion| room n2|
|#2|front room| piano | rooms s1,w3,e4|
|#3|library| spiders| rooms e2, n5|
|#4|kitchen| bats| rooms w2, n7|
|#5| dining room | dust, empty box| room s3|
|#6| vault | 3 walking skeletons | room e7|

|#8| secret room| piles of gold | room w6|

```python
#Text for relaying to user:

#Welcome to the haunted Castle Zork

rm1 = " \nYou start in the Foyer. There is a dead scorpion in the corner. It looks new.\nTo the (N) is a door"

rm2 = "\nThis is the front room, with a magnificent grand piano under layered dust. \nTo the (S) is a door \nTo the (W) is a door\nTo the (E) is a door"
    
rm3 = "\nThis is the beautiful library, with molding tomes and a giant sleeping spider!\nTo the (E) is a door\nTo the (N) is a door"

rm4 = "\nThis is the kitchen. The smell here is awful, likely from the bats sleeping on the cieling.\nTo the (W) is a door\nTo the (N) is a door"
    
rm5 = "\nThis is the formal dining room. All we have here is cobwebs, dust, and an an empty box.\nTo the (S) is a door"
    
rm6 = "\nThis is the vault. The door is very heavy...and there are 3 skeletons walking around! Run!\nTo the (E) is a door"

rm7 = " \nThis is the parlor. In the corner: a treasure chest! Sadly, it has long been looted. \nTo the (S) is a door\nTo the (W) is a door"

rm8 = "\nYou've found passage down to the Secret Treasure Room! Filled with treasure beyond wildest abandon!\nA trapdoor (t) is the only way out..."
```

```python
# Attempt No.2

loop = 1
print("-----------------------------------")
print("Welcome to Castle Zork\n Navigate around these haunted rooms, and try to find the secret treasure room!\n Controls are the cardinal directions: NSEW")
name = str(input("What is your name? "))
while True:  
      
    if loop == 1:
        print(rm1)
        dir1 = input()
        if dir1 == 'N':
            print(name +" walks north")
            loop = 2
        if dir1 == 'q':
            loop = 9 

    if loop == 2:
        print(rm2)
        dir2 = input()
        if dir2 == 'S':
            print(name + " goes back to the foyer...")
            loop = 1
        if dir2 == 'E':
            print(name + " walks east")
            loop = 4
        if dir2 == 'W':
            print(name + " walks west")
            loop = 3
        if dir2 == 'q':
            loop = 9 

    if loop == 3:
        print(rm3)
        dir3 = input()
        if dir3 == 'N':
            print(name + " walks north")
            loop = 5
        if dir3 == 'E':
            print(name + " goes back to the front room")
            loop = 2
        if dir3 == 'q':
            loop = 9    

    if loop == 4:
        print(rm4)
        dir4 = input()
        if dir4 == 'N':
            print(name + " walks north")
            loop = 7
        if dir4 == 'W':
            print(name + " returns to the front room")
            loop =2
        if dir4 == 'q':
            loop = 9 

    if loop == 5:
        print(rm5)
        dir5 = input()
        if dir5 == 'S':
            print(name + " goes back to the library")
            loop = 3
        if dir5 == 'q':
            loop = 9 

    if loop ==6:
        print(rm6)
        dir6 = input()
        if dir6 == 'E':
            print(name +" goes back to the parlour")
            loop = 7
        if dir6 == 'e':
            loop = 8
        if dir6 == 'q':
            loop = 9 

    if loop == 7:
        print(rm7)
        dir7 = input()
        if dir7 == 'W':
            print(name + " walks west")
            loop = 6
        if dir7 == 'S':
            print(name + " goes back to the kitchen")
            loop = 4
        if dir7 == 'q':
            loop = 9
            
    if loop == 8:
        print(rm8)
        dir8 = input()
        if dir8 == 't':
            print(name + " opens the trapdoor and laboriously climb up the chute...")
            loop = 4

    if loop == 9:
        print("Would " + name + " like to quit? y/n")
        dir9 = input()
        if dir9 == 'y':
            import random
            lower = 1
            upper = 3
            rnd = random.randint(lower, upper)
            if rnd == 1:
                print(name + " was suddenly eaten by a troll! RIP...")
                break
            if rnd == 2:
                print(name + " is too scared to continue, and runs out of the castle screaming into the distance, only screaming.")
                break
            if rnd == 3:
                print("Do you hear that, " + name + "? That sounds like music...Aa! The skeleton! My bones! Sudden onset Fibrodysplasia ossificans progressiva!")
                break
        if dir9 == 'n':
            print(name + "Goes back to the beginning then!")
            loop = 1
```
