````python
# getting the libraries we need
import urllib.request as urllib
````

````python
# setting the URL
wiki = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland" # URL for the page we are going to be lifting from.
````

````python
# creating a variable out of the parsed page
page = urllib.urlopen(wiki)
````

````python
#showing our variable
page
````

````python
# getting the Beautiful Soup parsing library | https://www.crummy.com/software/BeautifulSoup/bs4/doc/
from bs4 import BeautifulSoup
````

````python
soup = BeautifulSoup(page, "lxml")
````

````python
# this will print the webpage source with whitespace for human-readability
print(soup.prettify())
````

````python
# we can tell soup to print any HTML tags that we wish--as long as the page in question has such tags! It will print all content within the
# specified tag, so soup.body will return most all of the page!
print(soup.title)
````

````python
# soup.find will print and return all of a certain type (i.e. links in this case: <a href...> tags)
soup.find_all("a")
````

````python
all_links = soup.find_all("a") # creates a variable for the above list of all links
for link in all_links: # setting up a conditional
    print(link.get("href")) # prints only the href content of the link--the URL, rather than the class, link, title, and tags
````

````python
# looking for the table
soup.find_all("table")
````

````python
right_table = soup.find("table", class_="wikitable sortable") # making sure we get the lower table, with the correct content, and not a wrong one!
right_table # verifying that we have the right stuff... We do!
````

````python
AA = []
AB = []
AC = []
AD = []
AE = []
AF = []
AG = []
AH = []
AI = []
AJ = []
AK = []
AL = []



for row in right_table.findAll("tr"):   # the table actually is set TR > TH (County Names) > TD (data)
    cells = row.findAll("th")           # which made everything rather confusing!
    AA.append(cells[0].find(text=True)) # So we pulled TH (county Name) as AA, and broke out the TD data down below
    #print(len(cells))                   # 

    cells = row.findAll('td')                # here we get into the table data
    if len(cells) == 10:                     # the table is 10 cells 'wide' 
        AB.append(cells[0].find(text=True))  # start counting from 0, as always
        AC.append(cells[1].find(text=True))  #
        AD.append(cells[2].find(text=True))  #
        AE.append(cells[3].find(text=True))  #
        AF.append(cells[4].find(text=True))  #
        AG.append(cells[7].find(text=True))  # skipping rows 5 and 6 because those are photos and we don't want it
        AH.append(cells[8].find(text=True))  #
     #   AI.append(cells[8].find(text=True))
      #  AJ.append(cells[9].find(text=True))
       # AK.append(cells[10].find(text=True))
        #AL.append(cells[11].find(text=True))
        
AA.remove('County') # removing the extra entry    
print(AA) # we have an extra entry of 'County' from the first TH (of which the second TH is a part!) 
# now it's all clean!
````

````python
import pandas as pd

df = pd.DataFrame(AA, columns=['County']) # turning column AA into a data frame first

df['FIPS'] = AB
df['Est.'] = AC
df['County Seat'] = AD
df['Origin'] = AE
df['Etymology'] = AF
df['Population'] = AG
df['Area'] = AH

df.head()
````
