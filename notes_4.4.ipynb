```python
location = "./Module_0/datasets/gradedata.csv"
df = pd.read_csv(location)
df.head()
```

### Cleaning by Standard Deviation

```python
#still using gradedata.csv
#check df.head() if needed

#Standard Deviation Method

meangrade = df['grade'].mean()
stdgrade = df['grade'].std()
toprange = meangrade + stdgrade * 1.96  # why is it being done at 1.96 and not another number?
botrange = meangrade - stdgrade * 1.96  

# the forumla for such an activity is:  (Mean + 2*SD) and any points below (Mean - 2*SD) before plotting the frequencies.
# https://www.kdnuggets.com/2017/02/removing-outliers-standard-deviation-python.html

newdf = df.copy() #to not mess up the original df
newdf = newdf.drop(newdf[newdf['grade'] > toprange].index)
newdf = newdf.drop(newdf[newdf['grade'] < botrange].index)

newdf.head()
```
